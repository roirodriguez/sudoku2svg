```python
from sudoku2svg import SudokuBuilder, sudoku2svg

SUDOKU_STR = "1002000002100000"
sudoku = SudokuBuilder.fromString(SUDOKU_STR)
```

```python
svgdata = sudoku2svg(sudoku)
```

```python
from IPython.display import SVG, display, HTML
import base64
_html_template='<img width="{}" src="data:image/svg+xml;base64,{}" >'

def svg_to_fixed_width_html_image(svg, width="400px"):
    text = _html_template.format(width, base64.b64encode(svg).decode("utf=8"))
    return HTML(text)

svg_to_fixed_width_html_image(svgdata)
```

```python
sudoku.cells[4].value = 2
sudoku.cells[7].value = 1
sudoku.cells[13].value = 1
sudoku.cells[14].value = 2

for cell in sudoku.cells:
    if not cell.value:
        cell.center_marks = [3, 4]
```

```python
svgdata = sudoku2svg(sudoku)
svg_to_fixed_width_html_image(svgdata)
```
